```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
ls = pd.read_csv('AB_NYC_2019.csv')
```

```python
ls
```

```python
plt.hist(ls['price'], bins = np.arange(0,1000,40))
plt.xlabel('price(in US Dollars)')
plt.show()
```

```python
plt.scatter(x = ls['price'], y = ls['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('Number Of reviews')
plt.show()
```

```python
plt.scatter(x = ls['price'], y = ls['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('Number Of reviews')
plt.xlim(0,1100)
plt.show()
```

```python
plt.scatter(x = ls['price'], y = ls['number_of_reviews'], s=5)
plt.xlabel('price')
plt.ylabel('Number Of reviews')
plt.xlim(0,300)
plt.ylim(0, 310)
plt.show()
```
